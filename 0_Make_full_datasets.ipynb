# Process flux convergences

```python
path = "/Users/tatsumonkman/ecco.jpl.nasa.gov/drive/files/Version4/Release4/nctiles_monthly/"

variables = [
             "ADVr_SLT",
             "ADVr_TH",

             "ADVx_SLT",
             "ADVx_TH",

             "ADVy_SLT",
             "ADVy_TH",

             "DFrE_SLT",
             "DFrE_TH",

             "DFrI_SLT",
             "DFrI_TH",

             "DFxE_SLT",
             "DFxE_TH",    

             "DFyE_SLT",
             "DFyE_TH",    
            ]



for VAR in variables:
    VAR_years = open_datasets.open_ECCOv4r4_files(path,VAR)


    # make yearly mean dataset since WVELMASS is too big
    VAR_years_list = [var_da[VAR].mean(dim="time") for t, var_da in enumerate(VAR_years)]

    VAR_yearly_means = xr.concat(VAR_years[:3],dim="time").isel(time=np.arange(1,27)).assign_coords(time=np.arange(0,26))[VAR]*0

    for i, t in enumerate(VAR_yearly_means.time.values):
        print(t,end=" ")
        VAR_yearly_means.loc[{"time":t}] = VAR_years_list[i]

    VAR_yearly_means.to_netcdf(f"{VAR}_yearly.nc")
```
